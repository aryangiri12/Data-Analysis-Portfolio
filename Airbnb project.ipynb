```python
import pandas as pd

df = pd.read_csv('compressed_data.csv.gz', compression='gzip', encoding='ISO-8859-1')  # or try encoding='latin1'
```

```python
df.head()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.columns
```

# checking missing values

```python
print(df.isnull().sum())
```

# handling missing values

```python
df['last review'] =pd.to_datetime(df['last review'],errors = 'coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per months':0,'last review' :df['last review'].min()},inplace = True)
```

```python
df.dropna(subset ={'NAME','host name'},inplace = True)
```

```python
print(df.isnull().sum())
```

```python
df = df.drop(columns ={'license','house_rules'},errors ='ignore')
```

```python
df.head()
```

```python
# remove dollar signs and convert to float
import pandas as pd

# Clean and convert 'price' column
df['price'] = pd.to_numeric(df['price'].replace(r'[\$,]', '', regex=True), errors='coerce')

# Clean and convert 'service fee' column
df['service fee'] = pd.to_numeric(df['service fee'].replace(r'[\$,]', '', regex=True), errors='coerce')
```

```python
df.head()
```

# remove duplicates

```python
df.drop_duplicates(inplace =True)
```

```python
df.info()
```

# descriptive statistics

```python
df.describe()
```

#  Visualisation

# what is the distribution of listing prices ?

```python
plt.figure(figsize =(10,6))
sns.histplot(df['price'],bins =50,kde= True,color ="g")
plt.title('Distribution of listing price')
plt.xlabel("price $")
plt.ylabel("Frequency")
plt.show()
```

# How  are different room types distributed?

```python
df['room type']
```

```python
plt.figure(figsize =(8,5))
sns.countplot(x ='room type',data = df,color ="hotpink")
plt.title("room type distribution")
plt.xlabel("room type")
plt.ylabel('count')
plt.show()
```

# How listings distributes across different nighborhoods ?

```python
plt.figure(figsize =(12,8))
sns.countplot(y='neighbourhood group', data=df, color="lightgreen", order=df['neighbourhood group'].value_counts().index)

plt.title("Number of listing by negihborhood Group")
plt.xlabel('Count')
plt.ylabel("Neighborhood Group")
plt.show()
```

# What is the relationship between price and room type

```python
import seaborn as sns

plt.figure(figsize=(10,6))
sns.boxplot(x='room type', y='price', data=df, palette='Set1')
plt.title('Price vs Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.show()
```

# How has the number of reviews changes over with time

```python
df['last review'] = pd.to_datetime(df['last review'])
review_over_time = df.groupby(df['last review'].dt.to_period('M')).size()


plt.figure(figsize =(12,6))
review_over_time.plot(kind ='line',color ='red')
plt.title('Number of reviews over Time')
plt.xlabel("Date")
plt.ylabel("Number of Reviews")
plt.show()
```
